```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
# Load the dataset and select specific columns
df = pd.read_csv('train.csv')[['Age', 'Pclass','SibSp', 'Parch', 'Survived']]
```

```python
df.head(5)
```

```python
# this will drop all the rows with missing values in the dataset

df.dropna(inplace=True)
```

```python
df.head()
```

```python
# Split the dataset into features (X) and target variable (y)
#: (before the comma): Tells Pandas to select every single row in the dataset.
# 0:4 (after the comma): Tells Pandas to select columns starting at index 0 up to, but not including, index 4.The Result:
#  It grabs columns 0, 1, 2, and 3 across all rows and stores them in a new DataFrame called X.


X = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

### Appling Feature Construction

```python
X['Family_Size'] = X['SibSp'] + X['Parch']
```

```python
X.head()
```

```python
def myfunc(num):
    if num == 1:
        #alone
        return 0
    elif num >1 and num <=4:
        # small family
        return 1
    else:
        # large family
        return 2
```

```python
myfunc(4)

```

```python
X['Family_type'] = X['Family_size'].apply(myfunc)
X.head()
```

```python
X.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```
